## 问题 1

根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$

![问题1.jpg](./fig/问题1.jpg)

````python
import math
import sys
from functools import lru_cache

# 1. T(n) = T(n/4) + 1, T(1)=1
@lru_cache(maxsize=None)
def T1(n):
    if n == 1:
        return 1
    return T1(n // 4) + 1

# 2. T(n) = 3T(n/3) + n^2, T(1)=1
@lru_cache(maxsize=None)
def T2(n):
    if n == 1:
        return 1
    return 3 * T2(n // 3) + n * n

# 3. T(n) = T(2n/3) + 1, T(1)=1
@lru_cache(maxsize=None)
def T3(n):
    if n == 1:
        return 1
    return T3(2 * n // 3) + 1

# 4. T(n) = 5T(n/4) + n, T(1)=1
@lru_cache(maxsize=None)
def T4(n):
    if n == 1:
        return 1
    return 5 * T4(n // 4) + n

# 5. T(n) = 1 (n<=2); T(n) = T(sqrt(n)) + 1 (n>2)
@lru_cache(maxsize=None)
def T5(n):
    if n <= 2:
        return 1
    return T5(int(math.sqrt(n))) + 1

def asymptotic_T1(n): return math.log(n, 4)          # T1 = Θ(log n)
def asymptotic_T2(n): return n * n                   # T2 = Θ(n^2)
def asymptotic_T3(n): return math.log(n, 3/2)        # T3 = Θ(log n)
def asymptotic_T4(n): return n ** math.log(5, 4)     # T4 = Θ(n^{log_4 5})
def asymptotic_T5(n): return math.log2(math.log2(n)) if n > 2 else 0  # Θ(log log n)

def test_all():
    print("=" * 70)
    print("验证递归算法数值与理论渐近上界的关系（数值可能差常数倍）")
    print("=" * 70)

    test_points_14 = [4**k for k in range(1, 9)]
    test_points_2  = [3**k for k in range(1, 8)]
    test_points_3  = [int((3/2)**k * 10) for k in range(2, 9)]
    test_points_5  = [2**(2**k) for k in range(0, 5)]

    print("\n1) T(n)=T(n/4)+1   vs   log4(n)")
    for n in test_points_14:
        print(f"n={n:6d}, T1={T1(n):4d}, log4(n)={asymptotic_T1(n):.2f}")

    print("\n2) T(n)=3T(n/3)+n^2   vs   n^2")
    for n in test_points_2:
        print(f"n={n:5d}, T2={T2(n):8d}, n^2={asymptotic_T2(n):8.0f}")

    print("\n3) T(n)=T(2n/3)+1   vs   log_{3/2}(n)")
    for n in test_points_3:
        print(f"n={n:4d}, T3={T3(n):3d}, 理论值≈{asymptotic_T3(n):.2f}")

    print("\n4) T(n)=5T(n/4)+n   vs   n^{log_4(5)}")
    for n in test_points_14:
        t4 = T4(n)
        theory = asymptotic_T4(n)
        print(f"n={n:6d}, T4={t4:6d}, n^log={theory:6.1f}, 比值={t4/theory:.3f}")

    print("\n5) T(n)=T(sqrt(n))+1   vs   log2(log2(n))")
    for n in test_points_5:
        t5 = T5(n)
        theory = asymptotic_T5(n)
        print(f"n={n:6d}, T5={t5:3d}, log(log(n))={theory:.2f}")

if __name__ == "__main__":
    test_all()
````

````python
# add your code here
# your algorithm time complexity is:
import math

def do_something(A, p, r):
    n = r - p + 1
    if n == 2 and A[p] > A[r]:
        A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        do_something(A, p, p + m - 1)
        do_something(A, r - m + 1, r)
        do_something(A, p, p + m - 1)

def stooge_sort(A):
    if len(A) > 1:
        do_something(A, 0, len(A) - 1)

if __name__ == "__main__":
    test_cases = [
        [3, 1, 2],
        [7, 6, 5, 4, 3, 2, 1],
        [1, 3, 2, 4],
        [5, 1, 4, 2, 8, 0, 2],
        [9, -3, 5, 2, 6, 8, -6, 1, 3]
    ]
    for arr in test_cases:
        original = arr.copy()
        stooge_sort(arr)
        print(f"原始: {original}  ->  排序后: {arr}")
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

1.算法设计思路：
采用暴力枚举所有可能的买入日和卖出日组合。设数组 profits 长度为 n，索引从 0 到 n-1 对应第 1 天至第 n 天。
对每一对买入日 i（0 ≤ i ≤ n-2）和卖出日 j（i ≤ j ≤ n-1），计算区间 [i, j] 内收益的总和，该总和即为第 i+1 天买入、第 j+1 天卖出所能获得的利润。
用变量 max_profit 记录当前最大利润，初始可设为负无穷（或 0，若允许不交易）。若某组 (i, j) 的总和大于 max_profit，则更新最大利润及对应的买入日 buy = i、卖出日 sell = j。
遍历结束后，若 max_profit ≤ 0，说明无法获取正利润，可返回利润 0 且不给出买卖日期；否则返回 max_profit 以及 buy+1, sell+1（转换为题目中的第几天）。

时间复杂度：
两层循环共执行 O(n^2) 次，每次通过累加计算子数组和，故总时间复杂度为 O(n^2)。空间复杂度为 O(1)，仅使用常数个变量。

````python
# add your code here
# your algorithm time complexity is:
def max_profit_bruteforce(profits):
    n = len(profits)
    max_profit = float('-inf')
    buy = sell = None

    for i in range(n - 1):
        current_sum = 0
        for j in range(i, n):
            current_sum += profits[j]
            if current_sum > max_profit:
                max_profit = current_sum
                buy, sell = i, j

    if max_profit <= 0:
        return 0, None, None
    return max_profit, buy + 1, sell + 1


profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = max_profit_bruteforce(profits)
print(f"最大利润: {profit}, 第{buy_day}天买入, 第{sell_day}天卖出")
````

2.算法设计思路：
本问题等价于求最大子数组和，但要求子数组长度至少为 2（买入日和卖出日不能相同）。
设前缀和 pre[t] = profits[1] + ... + profits[t]，pre[0] = 0。
在第 j 天卖出、第 i 天买入（i < j）的利润为 pre[j] - pre[i-1]。
对于固定的卖出日 j，要使利润最大，只需让 pre[i-1] 在 i ≤ j-1（即 i-1 ≤ j-2）的前提下取最小值。
因此我们可以顺序扫描每一天，同时维护在 j-2 及之前的最小前缀和 min_pre 及其对应的索引（买入日前一天）。

具体流程：若 n < 2，无合法交易，返回利润 0。初始化 cur_pre = 0（当前前缀和，对应 pre[t]），min_pre = 0（pre[0]），min_idx = 0（min_pre 对应的前缀索引，即 i-1）。从第 1 天遍历到第 n 天（设 t = 1..n）：保留上一轮的前缀和 prev_pre = cur_pre，它等于 pre[t-1]。更新 cur_pre += profits[t-1]，得到 pre[t]。当 t ≥ 2 时，计算 profit = cur_pre - min_pre。若 profit > max_profit，更新答案：max_profit = profit, buy = min_idx + 1（因为 min_idx 是 i-1）, sell = t。将 prev_pre（即 pre[t-1]）与 min_pre 比较：若更小，则更新 min_pre = prev_pre，min_idx = t-1，供下次循环使用。遍历结束后，若 max_profit ≤ 0 则返回 (0, None, None)，否则返回 (max_profit, buy, sell)。正确性：通过维护 j-2 之前的最小前缀和，保证买入日严格在卖出日之前至少一天。

时间复杂度：
仅需一次线性扫描，每次操作 O(1)，总时间复杂度 O(n)。
空间复杂度 O(1)，只使用了常数个变量。

````python
def max_profit_one_pass(profits):
    n = len(profits)
    if n < 2:
        return 0, None, None

    max_profit = float('-inf')
    buy = sell = None
    min_pre = 0
    min_idx = 0
    cur_pre = 0

    for t in range(1, n + 1):
        prev_pre = cur_pre
        cur_pre += profits[t - 1]

        if t >= 2:
            profit = cur_pre - min_pre
            if profit > max_profit:
                max_profit = profit
                buy = min_idx + 1
                sell = t

        if prev_pre < min_pre:
            min_pre = prev_pre
            min_idx = t - 1

    if max_profit <= 0:
        return 0, None, None
    return max_profit, buy, sell


if __name__ == "__main__":
    profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    profit, buy_day, sell_day = max_profit_one_pass(profits)
    print(f"最大利润: {profit}, 第{buy_day}天买入, 第{sell_day}天卖出")
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

1.算法设计思路：
数组 A 满足 A[1] > A[2] 且 A[n-1] < A[n]，所有元素互不相同。局部极小定义为某个元素既小于其左边邻居又小于其右边邻居。由边界条件可知，A[1] 不是局部极小（因为它大于 A[2]），A[n] 也不是局部极小（因为它大于 A[n-1]）。然而，序列从 A[1] 到 A[2] 下降，从 A[n-1] 到 A[n] 上升，表明中间必然存在至少一个局部极小值。利用这个单调性，可以采用二分搜索策略在 O(log n) 时间内找到一个局部极小值。

具体做法：
维护搜索区间端点 low = 0, high = n-1（0‑based 索引）。当 low ≤ high 时，取中点 mid = (low + high) // 2。获取左右邻居值：若 mid > 0 则 left = A[mid-1]，否则left 为正无穷；若 mid < n-1 则 right = A[mid+1]，否则 right 为正无穷。如果 A[mid] 同时小于 left 和 right，则 A[mid] 即为局部极小，返回它。否则，如果 left < A[mid]（左侧更小），说明从 A[low] > A[low+1] 的下降趋势到 left < A[mid] 的抬升趋势，左半区间 [low, mid-1] 中必然存在局部极小，将 high 更新为 mid-1。若不满足上述条件，则必定 right < A[mid]，右半区间 [mid+1, high] 中必然存在局部极小，将 low 更新为 mid+1。每次迭代搜索范围减半，因此只需对数次比较即可找到结果。

时间复杂度：
递归关系为 T(n) = T(n/2) + O(1)，解得 T(n) = O(log n)。


````python
# add your code here
# your algorithm time complexity is:
def find_one_local_min(A):
    n = len(A)
    low, high = 0, n - 1
    while low <= high:
        mid = (low + high) // 2
        left_val = A[mid - 1] if mid > 0 else float('inf')
        right_val = A[mid + 1] if mid < n - 1 else float('inf')
        
        if A[mid] < left_val and A[mid] < right_val:
            return A[mid]
        if mid > 0 and A[mid - 1] < A[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return None

A = [9, 3, 7, 2, 1, 4, 5]
print("找到一个局部最小值:", find_one_local_min(A))
````

2.算法设计思路：
由定义，局部极小值必须严格小于左右相邻元素。由于数组满足 A[1]>A[2] 且 A[n−1]<A[n]，首尾元素不可能是局部极小，所有局部极小仅可能出现在下标 2 到 n−1 之间。要找出所有局部最小值，只需对数组进行一次线性扫描：依次检查每个内部位置 i，若 A[i]<A[i−1] 且 A[i]<A[i+1]，则将 A[i] 加入结果列表。扫描结束后返回结果列表。该算法简单直接，且能保证找到每一个局部极小值，因为最坏情况下局部极小值的个数可能达到 ⌊n/2⌋，任何算法都至少需要 Ω(n) 的时间才能输出所有结果。

时间复杂度：
仅需一次遍历，每个元素执行 O(1) 次比较，总时间复杂度为 Θ(n)。空间复杂度为 O(k)，其中 k 是局部极小值的数量，用于存储结果（不计输出空间则为 O(1) 额外空间）。

````python
def find_all_local_min(A):
    n = len(A)
    result = []
    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            result.append(A[i])
    return result


if __name__ == "__main__":
    A = [9, 3, 7, 2, 1, 4, 5, 1]
    print("所有局部最小值:", find_all_local_min(A))
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

1.算法设计思路：
采用高效的比较排序算法先将数组按升序完全排序，然后按顺序输出前 k 个元素。对于基于比较的排序，可选时间复杂度为 O(n log n) 的算法，如归并排序或堆排序。Python 内置的 sorted 函数使用 Timsort，平均和最坏时间复杂度均为 O(n log n)，适合直接使用。排序后，取前 k 个即可。若 k > n，则返回全部元素。

时间复杂度：
排序步骤为 O(n log n)，输出前 k 个为 O(k)，总时间复杂度 O(n log n + k) = O(n log n)（因为 k ≤ n）。空间复杂度取决于排序算法，Python sorted 返回新列表，需要 O(n) 额外空间。

````python
# add your code here
# your algorithm time complexity is:
def k_smallest_by_sorting(arr, k):
    if k <= 0:
        return []
    sorted_arr = sorted(arr)
    return sorted_arr[:k]

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(k_smallest_by_sorting(arr, k))
````

2.算法设计思路：
利用最小堆来不断取出最小值。建堆方式采用逐个插入：从空堆开始，将每个元素依次插入（heapq.heappush）。每次插入后维护堆性质，单次插入 O(log n)，共 n 个元素，因此建堆总时间复杂度为 O(n log n)。然后执行 k 次 Extract‑min 操作：使用 heapq.heappop 取出堆顶（当前最小值），将其加入结果列表。每次弹出也是 O(log n)，k 次共 O(k log n)。总时间复杂度 O((n + k) log n)。若 k 远小于 n，该方法不如“部分建堆 + 弹出”高效，但完全符合题目“不断插入建堆”的要求。

时间复杂度：
逐个插入建堆：O(n log n)。
k 次弹出最小值：O(k log n)。
总时间复杂度：O((n + k) log n)。
空间复杂度：需要 O(n) 额外空间存储堆。

````python
import heapq

def k_smallest_by_heap(arr, k):
    if k <= 0:
        return []
    heap = []
    for val in arr:
        heapq.heappush(heap, val)
    
    result = []
    for _ in range(min(k, len(heap))):
        result.append(heapq.heappop(heap))
    return result

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(k_smallest_by_heap(arr, k))
````

3.算法设计思路：
当数据总量超过内存限制、只能存储 k 个元素时，我们使用一个最大堆（容量固定为 k）在线处理数据流。初始时堆为空。依次读取每一条数据（数据可以来自磁盘文件、网络流等，无需全部载入内存）：若堆中元素数不足 k，直接将当前值插入堆中（维护最大堆性质）。若堆已满（大小 = k），比较当前值 x 与堆顶（当前堆中的最大值）：若 x 小于堆顶，则用 x 替换堆顶（弹出最大值，插入 x）；否则跳过。处理完所有数据后，堆中保存的就是全局最小的 k 个数。最后将堆中元素排序输出即可。由于 Python 标准库 heapq 实现的是最小堆，我们通过存储负值来模拟最大堆：堆顶的负数即为当前堆中真实的最大值。该算法每次只需要常数数量的比较和堆调整操作，空间固定在 O(k)，完全符合内存限制。

时间复杂度：
处理每条数据时最多进行一次堆替换操作（复杂度 O(logk)），因此总时间复杂度为 O(nlogk)。最后对堆内 k 个元素排序需要 O(klogk)，通常 k≪n，整体仍然是 O(nlogk)。

````python
import heapq

def top_k_smallest(stream, k):
    if k <= 0:
        return []

    max_heap = []

    for x in stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -x)
        else:
            if x < -max_heap[0]:
                heapq.heapreplace(max_heap, -x)

    result = [-val for val in max_heap]
    result.sort()
    return result

if __name__ == "__main__":
    import random
    large_data = (random.randint(1, 10**7) for _ in range(10**6))

    k = 10
    smallest_k = top_k_smallest(large_data, k)
    print("最小的", k, "个数:", smallest_k)

    test_arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
    print("\n验证示例数组:")
    print("arr =", test_arr)
    print("前3个最小:", top_k_smallest(test_arr, 3))
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

算法设计思路：
数组 A 按从左到右的顺序给出星球的初始位置，元素的符号代表运动方向（正向右，负向左），绝对值代表质量。所有星球速度相同，因此只有相向运动的星球才可能碰撞，具体表现为：左边的星球向右（正数）与右边的星球向左（负数）相遇时发生碰撞。碰撞规则：比较两者的绝对值（质量），质量小的消失；若质量相等，则向左运动的星球（负数）消失，向右的星球（正数）保留。由于同向星球永远不可能碰撞，且碰撞只发生在相邻的相向星球之间，可以采用栈来模拟整个过程。

遍历数组中的每个星球 x：如果 x > 0（向右），直接入栈。如果 x < 0（向左），则需要与栈顶的向右星球碰撞：当栈非空且栈顶元素 > 0 时，取栈顶 t：若 |t| > |x|：当前向左星球 x 被撞毁，不再入栈，结束碰撞。若 |t| < |x|：栈顶向右星球被撞毁，弹出栈顶，继续与下一个栈顶碰撞。若 |t| == |x|：质量相等，按规则向左星球消失、向右星球保留，即 x 被撞毁，栈中 t 保持不变，结束碰撞。如果碰撞结束后栈为空或栈顶为向左星球（负数），则当前向左星球 x 不会发生碰撞，将其入栈。遍历结束后，栈内元素自底向上即为从左到右最终存活的星球序列。

时间复杂度：
每个星球至多入栈一次、出栈一次，每次比较和弹出操作为 O(1)，因此总时间复杂度为 O(n)。空间复杂度为 O(n)，用于存储栈。

````python
# add your code here
# your algorithm time complexity is:
def planet_collision(A):
    stack = []
    for x in A:
        if x > 0:
            stack.append(x)
        else:  # x < 0
            while stack and stack[-1] > 0:
                t = stack[-1]
                if abs(t) > abs(x):
                    break
                elif abs(t) < abs(x):
                    stack.pop()
                else:
                    break
            else:
                stack.append(x)
                continue
    return stack

A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
print(planet_collision(A1))

A2 = [23, -8, 9, -3, -7, 9, -23, 22]
print(planet_collision(A2))
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

算法设计思路：
采用动态规划求解 0-1 背包。定义 dp[i][w] 表示考虑前 i 件物品，在背包容量为 w 时的最大价值。

状态转移方程：若不选第 i 件物品：dp[i][w] = dp[i-1][w]；若选第 i 件物品（需 w >= weight[i]）：dp[i][w] = max(dp[i-1][w], dp[i-1][w - weight[i]] + value[i])。初始化 dp[0][w] = 0（没有物品时价值为0）。依次填写表格，最终答案在 dp[n][W]，其中 n=4, W=9。也可以使用空间优化的一维数组，从后向前更新。

时间复杂度：
状态数量为 (n+1) × (W+1)，每个状态 O(1) 转移，因此时间复杂度为 O(nW)。本题 n=4, W=9，非常小。
空间复杂度：二维数组 O(nW)；若优化为一维数组则为 O(W)。

````python
# add your code here
# your algorithm time complexity is:
def knapsack_01(weights, values, capacity):
    n = len(weights)
    dp = [0] * (capacity + 1)
    for i in range(n):
        wgt = weights[i]
        val = values[i]
        for w in range(capacity, wgt - 1, -1):
            if dp[w - wgt] + val > dp[w]:
                dp[w] = dp[w - wgt] + val
    return dp[capacity]

weights = [5, 4, 6, 3]
values = [10, 40, 30, 50]
capacity = 9

max_value = knapsack_01(weights, values, capacity)
print("最大价值为:", max_value)
````

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

算法设计思路：
一棵二叉搜索树的中序遍历序列应该是严格递增的。现在恰好有两个节点的值被错误交换，导致中序序列中出现逆序。通过中序遍历找出这两个错误节点，然后交换它们的值即可恢复 BST，不改变树的结构。

具体步骤：对给定的 BST 执行中序遍历，在遍历过程中维护一个前驱节点指针 prev。比较当前节点与 prev 的值：如果 prev.val > curr.val，说明出现了逆序。第一次遇到逆序时，记录 first = prev。每次遇到逆序时，记录 second = curr（若整个遍历只有一处逆序，则这两个节点相邻；若两处，则第二次逆序时更新 second 为第二个更小的节点）。遍历结束后，交换 first 和 second 节点的值。

时间复杂度：
中序遍历访问每个节点一次，时间复杂度为 O(n)，其中 n 是二叉树节点总数。空间复杂度取决于递归栈深度，平衡时为 O(log n)，最坏为 O(n)；若采用 Morris 遍历可优化至 O(1)，但递归实现已足够清晰。

````python
# add your code here
# your algorithm time complexity is:
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(nums):
    if not nums or nums[0] is None:
        return None
    root = TreeNode(nums[0])
    queue = deque([root])
    i = 1
    while queue and i < len(nums):
        node = queue.popleft()
        if i < len(nums) and nums[i] is not None:
            node.left = TreeNode(nums[i])
            queue.append(node.left)
        i += 1
        if i < len(nums) and nums[i] is not None:
            node.right = TreeNode(nums[i])
            queue.append(node.right)
        i += 1
    return root

def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

def recoverTree(root):
    first = second = prev = None

    def inorder(node):
        nonlocal first, second, prev
        if not node:
            return
        inorder(node.left)
        if prev and prev.val > node.val:
            if not first:
                first = prev
            second = node
        prev = node
        inorder(node.right)

    inorder(root)
    if first and second:
        first.val, second.val = second.val, first.val

print("示例 1:")
input1 = [1, 3, None, None, 2]
print("输入: root =", input1)
root1 = build_tree(input1)
recoverTree(root1)
output1 = tree_to_list(root1)
print("输出:", output1)

print("示例 2:")
input2 = [3, 1, 4, None, None, 2]
print("输入: root =", input2)
root2 = build_tree(input2)
recoverTree(root2)
output2 = tree_to_list(root2)
print("输出:", output2)
````

## 问题 9 \*
给定 `n` 个非负整数表示每个宽度为 `1` 的柱子的高度图，计算按此排列的柱子，下雨之后能接多少雨水。



<div style="text-align:center;width:800px;"> <img alt="" src="./fig/hw3q6.png"> </div>

请分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

算法设计思路：
这个问题要求计算柱子间隙中能积存的雨水总量。每根柱子上方能积水的量取决于其左右两侧最高柱子的较小值减去自身高度（若正值）。我们可以利用双指针技巧在 O(n) 时间和 O(1) 空间内完成计算，无需额外数组。核心思想：用两个指针 left 和 right 分别从数组两端向中间移动。维护 left_max 记录从左侧开始到 left 位置遇到的最大高度，right_max 记录从右侧开始到 right 位置遇到的最大高度。如果 height[left] < height[right]，那么 left 位置的积水量由 left_max 决定：因为右侧至少有一个 height[right] 更高，所以木桶短板在左侧。若 height[left] < left_max，则积水量增加 left_max - height[left]；否则更新 left_max。然后 left 指针右移。否则，right 位置的积水量由 right_max 决定：左侧至少有一个更高的柱子，积水量增加 right_max - height[right]（若当前高度小于 right_max），否则更新 right_max。然后 right 指针左移。这样只需一次遍历就能累加所有积水量。

时间复杂度：
每个元素被访问一次，时间复杂度为 O(n)，其中 n 为数组长度。空间复杂度为 O(1)，只使用了常数个变量。

````python
# add your code here
# your algorithm time complexity is:
def trap(height):
    if not height:
        return 0

    left, right = 0, len(height) - 1
    left_max = right_max = 0
    water = 0

    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                water += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                water += right_max - height[right]
            right -= 1

    return water

if __name__ == "__main__":
    print(trap([0,1,0,2,1,0,1,3,2,1,2,1]))
    print(trap([4,2,0,3,2,5]))
````

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

 输入：root = [1,2,3], 输出：25

2.

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:

算法设计思路：
采用深度优先搜索（DFS）从根节点向下遍历。维护当前路径累积的数值 cur = cur * 10 + node.val。若当前节点是叶子（左右子节点均为空），则将 cur 加入总和。否则递归处理左、右子树，将当前累积值向下传递。整个过程每个节点访问一次，累加所有根到叶子路径生成的数字。

时间复杂度：
每个节点被访问一次，时间 O(n)（n 为节点总数）。
空间复杂度取决于递归栈深度，平衡树 O(log n)，最坏 O(n)。

````python
# add your code here
# your algorithm time complexity is:
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(nums):
    if not nums or nums[0] is None:
        return None
    root = TreeNode(nums[0])
    queue = deque([root])
    i = 1
    while queue and i < len(nums):
        node = queue.popleft()
        if i < len(nums) and nums[i] is not None:
            node.left = TreeNode(nums[i])
            queue.append(node.left)
        i += 1
        if i < len(nums) and nums[i] is not None:
            node.right = TreeNode(nums[i])
            queue.append(node.right)
        i += 1
    return root

def sumNumbers(root):
    total = 0

    def dfs(node, cur):
        nonlocal total
        if not node:
            return
        cur = cur * 10 + node.val
        if not node.left and not node.right:
            total += cur
            return
        dfs(node.left, cur)
        dfs(node.right, cur)

    dfs(root, 0)
    return total

root1 = build_tree([1, 2, 3])
print("示例 1 输出:", sumNumbers(root1))

root2 = build_tree([4, 9, 0, 5, 1])
print("示例 2 输出:", sumNumbers(root2))
````
